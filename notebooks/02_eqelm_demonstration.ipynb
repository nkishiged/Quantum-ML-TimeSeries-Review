```python
print("üî¨ D√©monstration de l'EQELM (Emulated Quantum Extreme Learning Machine)")
print("Impl√©mentation bas√©e sur la litt√©rature analys√©e dans notre revue")

import numpy as np
import matplotlib.pyplot as plt
from implementations.qelm.eqelm_implementation import EQELM
```

```python
print("üìä Chargement des donn√©es synth√©tiques...")
from data.synthetic.generate_data import generate_mackey_glass
```

```python
data = generate_mackey_glass(2000)
print(f"Donn√©es g√©n√©r√©es: {len(data)} points")
```

```python
window_size = 15
X, y = [], []

for i in range(len(data) - window_size):
    X.append(data[i:i+window_size])
    y.append(data[i+window_size])

X, y = np.array(X), np.array(y)
print(f"√âchantillons cr√©√©s: {len(X)} fen√™tres de taille {window_size}")
```

```python
print("üß† Entra√Ænement de l'EQELM...")
model = EQELM(n_qubits=6, n_features=window_size)
model.fit(X, y)

predictions = model.predict(X)
mse = np.mean((predictions - y)**2)
print(f"‚úì EQELM entra√Æn√© - MSE: {mse:.6f}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y[:200], label='Valeurs r√©elles', alpha=0.7)
plt.plot(predictions[:200], label='Pr√©dictions EQELM', alpha=0.7)
plt.xlabel('Temps')
plt.ylabel('Valeur')
plt.title('Pr√©dictions EQELM sur s√©rie Mackey-Glass')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('../benchmarks/results/eqelm_predictions.png', dpi=150, bbox_inches='tight')
plt.show()
```
